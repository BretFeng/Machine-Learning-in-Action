```python
import logisticRegression
import numpy as np
import matplotlib.pyplot as plt

parameterMatrix, category = logisticRegression.loadDataSet('testSet.txt')
# results for gradient descent
weightsGD, weightsHistoryGD = logisticRegression.gradientDescent(parameterMatrix, category, 0.1, 500)
print(weightsGD)
```

```python
iterNumber = np.arange(1, len(weightsHistoryGD[0])+1, 1)
fig = plt.figure()
ax = fig.add_subplot(311)             # 将画面分为3行1列，从左到右，从上到下的第1幅图
ax.plot(iterNumber, weightsHistoryGD[0])
bx = fig.add_subplot(312)             # 将画面分为3行1列，从左到右，从上到下的第2幅图
bx.plot(iterNumber, weightsHistoryGD[1])
cx = fig.add_subplot(313)             # 将画面分为3行1列，从左到右，从上到下的第3幅图
cx.plot(iterNumber, weightsHistoryGD[2])
plt.show()
```

```python
# results for stochastic gradient descent
weightsSGD, weightsHistorySGD = logisticRegression.stocGradientDescent(parameterMatrix, category, 0.1, 500)  
print(weightsSGD)
```

```python
iterNumber = np.arange(1, len(weightsHistorySGD[0])+1, 1)
fig = plt.figure()
ax = fig.add_subplot(311)             # 将画面分为3行1列，从左到右，从上到下的第1幅图
ax.plot(iterNumber, weightsHistorySGD[0])
bx = fig.add_subplot(312)             # 将画面分为3行1列，从左到右，从上到下的第2幅图
bx.plot(iterNumber, weightsHistorySGD[1])
cx = fig.add_subplot(313)             # 将画面分为3行1列，从左到右，从上到下的第3幅图
cx.plot(iterNumber, weightsHistorySGD[2])
plt.show()
```

```python
parameterMatrix, category = logisticRegression.loadDataSet('testSet.txt')
positiveX1 = []
positiveX2 = []
negativeX1 = []
negativeX2 = []
for i in range(len(category)):
    if category[i] == 1:
        positiveX1.append(parameterMatrix[i][1])
        positiveX2.append(parameterMatrix[i][2])
    elif category[i] == 0:
        negativeX1.append(parameterMatrix[i][1])
        negativeX2.append(parameterMatrix[i][2])

fig = plt.figure()

ax = fig.add_subplot(121)             # 将画面分为1行2列，从左到右，从上到下的第1幅图
ax.scatter(positiveX1, positiveX2, s = 30, c = 'red', marker = 's')
ax.scatter(negativeX1, negativeX2, s = 30, c = 'green')
x1 = np.arange(-3, 3, 0.1)
x2 = (-weightsGD[1] * x1 - weightsGD[0]) / weightsGD[2]   # 边界条件是w0*x0+w1*x1+w2*x2=0 (x0=1), 因为当这一串乘积为0时，sigmoid的输入为0，输出为0.5，即判定的临界条件 
ax.plot(x1, x2)
plt.xlabel('X1')
plt.ylabel('X2')

bx = fig.add_subplot(122)             # 将画面分为1行2列，从左到右，从上到下的第2幅图
bx.scatter(positiveX1, positiveX2, s = 30, c = 'red', marker = 's')
bx.scatter(negativeX1, negativeX2, s = 30, c = 'green')
x1 = np.arange(-3, 3, 0.1)
x2 = (-weightsSGD[1] * x1 - weightsSGD[0]) / weightsSGD[2]   # 边界条件是w0*x0+w1*x1+w2*x2=0 (x0=1), 因为当这一串乘积为0时，sigmoid的输入为0，输出为0.5，即判定的临界条件 
bx.plot(x1, x2)
plt.xlabel('X1')
plt.ylabel('X2')

plt.show()
```
