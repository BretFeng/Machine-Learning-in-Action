```python
import numpy as np
import matplotlib.pyplot as plt
import kNN
```

```python
datingDataMatrix, datingDataLabels = kNN.file2matrix('datingTestSet.txt')
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(datingDataMatrix[:,1], datingDataMatrix[:,2])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(datingDataMatrix.shape[0]):
    if datingDataLabels[i] == 'didntLike':
        ax.scatter(datingDataMatrix[i,0], datingDataMatrix[i,1], c = 'b')
    elif datingDataLabels[i] == 'smallDoses':
        ax.scatter(datingDataMatrix[i,0], datingDataMatrix[i,1], c = 'g')
    elif datingDataLabels[i] == 'largeDoses':
        ax.scatter(datingDataMatrix[i,0], datingDataMatrix[i,1], c = 'r')
plt.xlabel('annual flying distance')
plt.ylabel('time spending in video games %')
plt.show()
```

```python
normDataSet, ranges, minValueMatrix = kNN.normalization(datingDataMatrix)
print(normDataSet)
```
